```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

### Part A

```python
df = pd.read_csv("PCAData.csv", index_col = 0)

fig, ax = plt.subplots(figsize=(16, 10))
fig.suptitle("Scatter Plot of PCA Data", fontsize=20)
ax.axis('equal')
ax.axis('off')
plt.scatter(df['0'], df['1'])
```

The data in the scatterplot appears to show a football shape. The data shows a positive trend; they vary the most from the bottom left to the top right.

### Part B

```python
pca = PCA(n_components=2)
pca.fit(df)

print("Eigenvalues: ", np.round(pca.explained_variance_, 2))
print("Eigenvectors: ", np.round(pca.components_, 2))
```

The eigenpairs are as follows:

<center>

$\lambda _1 = 8.17$

<br>

$$E_1 = \begin{bmatrix}
-0.81 \\
-0.58 \\
\end{bmatrix}$$

<br>

$\lambda _2 = 1.73$

<br>

$$E_2 = \begin{bmatrix}
-0.58 \\
0.81 \\
\end{bmatrix}$$

</center>

<br>

### Part C

```python
x_mean = df['0'].mean()
y_mean = df['1'].mean()

arrow_1_termination = np.sqrt(pca.explained_variance_[0]) * pca.components_[0]
arrow_2_termination = np.sqrt(pca.explained_variance_[1]) * pca.components_[1]

fig, ax = plt.subplots(figsize=(16, 10))
fig.suptitle("Scatter Plot with Component Arrows", fontsize=20)
ax.axis('equal')
ax.axis('off')
plt.scatter(df['0'], df['1'])
plt.arrow(x_mean, y_mean, arrow_1_termination[0], arrow_1_termination[1], color='r', width=0.15)
plt.arrow(x_mean, y_mean, arrow_2_termination[0], arrow_2_termination[1], color='b', width=0.15)
```

The directions of the eigenvectors are orthogonal to each other. The first eigenvector points parallel to the largest variance component. It is pointing from the origin to down and to the left. The role of the vectors is to indicate both the strength of the variation of a component as well as to indicate a positive or negative correlation and what other components are orthogonal or not.
